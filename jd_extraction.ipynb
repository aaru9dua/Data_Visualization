# Function to extract skills

```python
nlp = spacy.load("en_core_web_lg")
def extract_jd(job_description):
    try:
        # init skill extractor
        skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)
        annotations = skill_extractor.annotate(job_description)
        doc_node_values = list(set([entry['doc_node_value'] for entry in annotations['results']['full_matches']]))+list(set([entry['doc_node_value'] for entry in annotations['results']['ngram_scored']]))
        return doc_node_values
    except Exception as e:
        print(f"Error: {e} \n {job_description}")
```

```python
df = pd.read_csv('Scraped_Indeed_Data/Data_Analyst_full__25.04.23.csv')
df.head()
```

```python
df.info()
```

```python
df.dropna( subset = ["job_description"], inplace=True)
```

```python
df.info()
```

```python
df['skills'] = df['job_description'].apply(lambda x: extract_jd(x))
```
