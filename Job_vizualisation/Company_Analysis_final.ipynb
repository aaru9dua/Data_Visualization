```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from collections import Counter
from PIL import Image

# Use Counter to count occurrences
employer_counts = Counter(list(map_data['Employer']))


# Sort the counts and extract data for plotting
sorted_counts = sorted(employer_counts.items(), key=lambda x: x[1], reverse=True)
names, counts = zip(*sorted_counts[:5])
print(names)


# Create a bar plot using Matplotlib
plt.figure(figsize=(10, 12))
bars = plt.bar(names, counts)

# Function to add logos on top of each bar
def add_logo_on_bar(logo_path, bar, ax):
    img = Image.open(logo_path)
    imagebox = OffsetImage(img, zoom=0.08)  # Adjust zoom as needed
    ab = AnnotationBbox(imagebox, (bar.get_x() + bar.get_width() / 2, bar.get_height()), frameon=False, pad=2)
    ax.add_artist(ab)

# Add logos on top of each bar
for bar, company_name in zip(bars, names):
    if company_name in logo_paths:
        logo_path = logo_paths[company_name]
        add_logo_on_bar(logo_path, bar, plt.gca())

# Set labels and title
plt.xlabel('Employer')
plt.ylabel('Count')
plt.title('Top Employers Across All Roles')

# Remove x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
plt.savefig('all_role_employers.png')
```
